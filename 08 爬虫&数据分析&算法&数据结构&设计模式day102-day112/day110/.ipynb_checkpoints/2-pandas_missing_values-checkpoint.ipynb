- DataFrame索引和切片的操作

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

df = DataFrame(data=np.random.randint(0,100,size=(8,6)))
df
```

```python
#索引
df[0] #df[列索引]
df[[1,2,3]] #df[[col1,col2,col3]] #批量取出多列
df.iloc[0] #df.iloc[行索引]
df.iloc[[1,2]] #df.iloc[[index1,index2]]#批量取出多行
df.iloc[0,1]#df.iloc[index,col] 取得指定元素
df.iloc[[1,2],3]#df.iloc[[index1,index2],col] #批量取出多个元素
```

```python
#切片
df[0:3] #df[行切片]
df.iloc[:,0:3] #df.iloc[行切片,列切片]
```

# 处理丢失数据

有两种丢失数据：
- None
- np.nan(NaN)

## 1. None

None是Python自带的，其类型为python object。因此，None不能参与到任何计算中。

```python
#查看None的数据类型
None + 1
```

## 2. np.nan（NaN）

np.nan是浮点类型，能参与到计算中。但计算的结果总是NaN。

```python
#查看np.nan的数据类型
np.nan + 1
```

## 3. pandas中的None与NaN

### 1) pandas中None与np.nan都视作np.nan

创建DataFrame

```python
df.iloc[2,4] = None
df.iloc[3,3] = np.nan
df.iloc[5,2] = None
df.iloc[2,1] = None

df
```

### 2) pandas处理空值操作

- ``isnull()``
- ``notnull()``
- ``dropna()``: 过滤丢失数据
- ``fillna()``: 填充丢失数据

(1)判断函数
- ``isnull()``
- ``notnull()``

```python
df
```

```python
df.isnull()
```

```python
df.loc[[True,True,False,False,True,False,True,True]]
```

- df.notnull/isnull().any()/all()

```python
df.notnull()
```

```python
df
```

```python
df.notnull().any(axis=1)
df.notnull().all(axis=1)
```

```python
df.loc[df.notnull().all(axis=1)]
```

```python
df.isnull().any(axis=1)
```

```python
#固定搭配
isnull=》any
notnull=》all
#结论：将df.notnull().all(axis=1)作为源数据的行索引，就可以将空对应的行删除
```

```python
indexs = df.loc[df.isnull().any(axis=1)].index #获取的是值对应行的行索引
df.drop(labels=indexs,axis=0)
```

df.dropna() 可以选择过滤的是行还是列（默认为行）:axis中0表示行，1表示的列

```python
df.dropna(axis=0)
```

(3) 填充函数 Series/DataFrame
- ``fillna()``:value和method参数

```python
df.fillna(method='bfill',axis=0)
```

可以选择前向填充还是后向填充
